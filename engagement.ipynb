```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
engage = pd.read_csv(r'C:\\Users\Servet\Desktop\takehome_user_engagement.csv')
engage = engage.set_index(pd.DatetimeIndex(engage['time_stamp']))
users = pd.read_csv(r'C:\\Users\Servet\Desktop\takehome_users.csv',encoding='latin-1')
users.head()
```

```python
engage.head()
```

```python
#groups by user, then date by week, and sums number of visits per week
user_id = engage.groupby('user_id').resample('1W').sum()
#removes weeks that were imputed for grouping but are NaN
user_id = user_id[user_id['user_id'].isnull()==False].drop('user_id',axis=1)
user_id.head(20)
```

```python
#tags weeks that had less than 3 visits for all users#tags w 
required_users = user_id['visited']>=3
#Groups by users and sums up the number of weeks that had 3 or more visits
required_users = required_users.groupby(level=0).apply(np.sum)
#removes users with no weeks that had 3 or more visits
required_users = required_users[required_users!=0]
required_users = pd.DataFrame(required_users,index=required_users.index)
required_users.columns = ['Active Weeks']
required_users.head(5)
```

```python
user_data = required_users.merge(users,left_index=True,right_on='object_id',how='inner')
user_data.head()
```

```python
#Pull data on Active users
user_data = required_users.merge(users,left_index=True,right_on='object_id',how='inner')
user_data['Active']=1

#Add Active column to origional Data
users['Active']=int(0)
users.iloc[list(user_data.index),10]=1
model_data = users.set_index('object_id').sort_index()

#Prep data for modeling by removing unecessary features
model_data = model_data.drop([ 'name', 'email','last_session_creation_time',],axis=1)
date_feature = model_data.pop('creation_time')
model_data.head()
```

```python
model_data['invited_by_user_id'] = model_data['invited_by_user_id'].fillna(0)
model_data['invited_by_user_id'] = model_data['invited_by_user_id'].apply(
        lambda x: 1 if x!=0 else 0)
model_data = pd.get_dummies(data=model_data,columns=['creation_source'])
```

```python
org = model_data.groupby('org_id').sum()
plt.scatter(org.index,org.Active)
plt.ylabel('# of Active users in Group')
plt.xlabel('Org ID')
plt.show()
```

```python
org_cata = []
for i in model_data['org_id']:
    cata = org.loc[i,'Active']
    org_cata.append(cata)
model_data['org_id'] = org_cata

model_data = pd.get_dummies(data=model_data,columns=['org_id'])
model_data.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV,train_test_split

y = model_data.pop('Active')
x = model_data
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.3,random_state=38)
```

```python
gbt = GradientBoostingClassifier(max_features='sqrt')
p = {'n_estimators':[40,45,50,55,60],'learning_rate':[.01,0.02,0.03,0.04,0.05],
              'max_depth':[1,2,3,4,5]}
grid1 = GridSearchCV(estimator=gbt,param_grid=p,
                     scoring='accuracy',cv=5)
grid1.fit(x_train,y_train)
best_para_gra = grid1.best_params_
best_acc_gra = grid1.best_score_

print('The Tuned Paratmers :\n',best_para_gra,'\nAchieved %s Percent Accuracy' %(best_acc_gra*100))
```
